Depois do git instalado na máquina, para verificar a versão basta digitar no terminal
      
    $ git --version

É necessário comunicar ao git quem é você, senão ele não será capaz de fazer as comunicações. Para gerar a lista de configurações do git, digite

    $ git config --global --list

Para inserir a configuração de um usuário no git digite

    $ git config --global user.name "CaikeExemplo"

Para testar se deu certo, basta entrar com o comando anterior com --list

Para inserir o email, basta digitar

    $ git config --global user.email caikedamke@gmail.com

Com essas configurações acabamos de nos identificar para o git quem sou eu.

> Alguns comandos úteis no terminal

    ls -> lista os objetos de uma pasta
    cd -> navega para as pastas
    cd .. -> volta na pasta anterior
    touch nome_do_arquivo.txt -> cria um arquivo no formato .txt, mas pode criar em qualquer formato desejado
    gedit nome_do_arquivo.txt ->  edita o arquivo
    nano nome_do_arquivo.txt -> faz a mesma coisa que o gedit
    cat nome_do _arquivo.txt -> mostra o que tem escrito no arquivo
    mkdir nome_diretorio -> cria um diretório
    rmdir nome_diretorio -> remove o diretório
    mv diretorio1 diretorio2 -> move o diretorio1 para o diretorio2. Serve para renomear também

Para inicializar o git em um repositório, basta digitar o comando que se segue dentro da pasta desejada

    $ git init

Para listar a pasta .git criada e posteriormente acessá-la, basta digitar

    $ ls -a
    $ cd .git

Para identificar os arquivos que estão ou não estão no stage, basta digitar

    $ git status
    On branch master #indica em que branch você está

    No commits yet #indica que não houve commits ainda

    Untracked files: #indica os arquivos não rastreados
      (use "git add <file>..." to include in what will be committed)
            index.html

    nothing added to commit but untracked files present (use "git add" to track)

Para rastrear o arquivo, basta digitar

    $ git add index.html

Para verificar o status

    $ git status
    On branch master

    No commits yet

    Changes to be committed: #mudanças prontas para serem comitadas
      (use "git rm --cached <file>..." to unstage)
        new file:   index.html

Para retirar um arquivo do stage, digite

    $ git rm --cached index.html

Se tem vários arquivos para serem adicionados e rastreados, para fazer de uma vez, basta digitar

    $ git add .

Para comitar um arquivo

    $ git commit -m "Adicionado arquivo index"

Para rever (histórico) os commits feitos

    $ git log
    #forma condensada
    $ git log --oneline

Para executar vários comandos de uma vez, basta usar o &&. Por exemplo, para adicionar o arquivo e comitar, basta fazer

    $git add . && git commit -m "Mensagem desejada"


Para fazer checkout em um commit. Ele vai voltar no tempo, nas alterações que você fez no arquivo que está trabalhando

    $ git checkout id_commit
    #Para voltar basta digitar
    $ git checkout nome_do_branch

Para reverter o histórico do commit
    
    $ git revert id_commit

Para resetar o commit, porém deleta todos os commits para a frente

    $ git reset id_commit



Criar um arquivo de ambiente

    $ touch .env

Dizer para o git não rastrear esse arquivo, primeiro cria o seguinte arquivo

    $ touch .gitignore

e depois vai até ele e insira o arquivo que deseja ser ignorado, neste caso abra o .gitignore, digite .env e salve. Depois é só dar o comit dizendo que o arquivo ignore goi adicionado

Podemos fazer isso para diretórios também.
